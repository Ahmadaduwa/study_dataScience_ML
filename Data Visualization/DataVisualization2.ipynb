# Workshop1: dowlaod and check

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

url = 'treatment-nl.csv'
df = pd.read_csv(url)

df.sample(5)
```

```python
df.info()
```

```python
df.isnull().sum() #check missing values
```

```python
df.describe().round(3) #check summary statistics
```

# Workshop2: Plot a graph to see the distribution.

```python
sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 14})
sns.distplot(df['product'], color='b')
plt.show()
```

# Workshop3: plot Scaater to see relationship

```python
sns.scatterplot(x="TB", y="product", data=df)
```

```python
sns.scatterplot(x="TC", y="product", data=df)
```

# Workshop4: Scatter plot in single graph

```python
sns.scatterplot(x="TA", y="product", data=df, label="TA")
sns.scatterplot(x="TB", y="product", data=df, label="TB")
sns.scatterplot(x="TC", y="product", data=df, label="TC")
plt.legend(loc='lower right')
```

# Workshop5: Use Subplot with Scatter

```python
f, axes = plt.subplots(1, 3, figsize=(10, 3))
sns.scatterplot(x="TA", y="product", data=df, ax=axes[0])
sns.scatterplot(x="TB", y="product", data=df, ax=axes[1])
sns.scatterplot(x="TC", y="product", data=df, ax=axes[2])
```

```python
f,axes = plt.subplots(2, 2, figsize=(8, 8))
sns.scatterplot(x="TA", y="product", data=df, ax=axes[0, 0])
sns.scatterplot(x="TB", y="product", data=df, ax=axes[0, 1])
sns.scatterplot(x="TC", y="product", data=df, ax=axes[1, 0])
sns.distplot(df['product'], ax=axes[1, 1], color='b')
```

# Workshop10: Heatmap

```python
corr_val = df.select_dtypes(include='number').corr()
sns.heatmap(corr_val, annot=True, cmap='coolwarm', center=0, fmt='.3f')
# 1 = Positive relationship, -1 = Negative relationship, 0 = No relationship
```
