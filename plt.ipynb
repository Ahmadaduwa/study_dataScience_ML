```python
year = 1
area = 30
income = 4500 * area

server = 100000 
electric_server = 30000
count = 1
labels = []
values = []
new_labels = []

while True:
    if area == 200:  # ออกจากลูปเมื่อพื้นที่เท่ากับ 200
        break
    
    if area % 30 == 10:   
        count += 1

    cost = electric_server * count
    
    if area % 30 == 10 or year == 1:
        cost += server

    income = 4500 * area
    profit = income - cost
    print("ปี:", year, "พื้นที่:", area, "รายได้:", income, "ต้นทุน:", cost, "กำไร:", profit)
    
    label_x = str(year) + "\n" + str(area)
    labels.append(int(year))
    values.append(profit)
    new_labels.append(label_x)

    year += 1
    area += 10  # เพิ่มพื้นที่ทีละ 10
```

```python
print(values)
```

```python
# สร้างกราฟแท่ง
plt.bar(labels, values, color='skyblue')

plt.plot(labels, values, marker='o', color='orange', linestyle='-', linewidth=2)

# ตั้งค่าชื่อและแกน
plt.title('Annual profit')
plt.xlabel('Year')
plt.ylabel('Profit')

plt.xticks(ticks=labels)
# แสดงกราฟ
plt.show()
```

```python

import matplotlib.pyplot as plt



# สร้างกราฟแท่ง
plt.bar(labels, values, color='skyblue')

# สร้างกราฟเส้นพร้อมจุด
plt.plot(labels, values, marker='o', color='orange', linestyle='-', linewidth=2)

# ตั้งค่าชื่อและแกน
plt.title('Annual profit')
plt.xlabel('Year')
plt.ylabel('Profit')

# ปรับการตั้งค่า xticks เพื่อให้แท็กไม่ทับกัน
plt.xticks(ticks=range(len(labels)), labels=new_labels, rotation=0, ha="right", fontsize=10)

# ใช้ tight_layout เพื่อปรับระยะห่างอัตโนมัติ
plt.tight_layout()

# แสดงกราฟ
plt.show()

```
