```python
import numpy as np
import matplotlib.pyplot as plt

# สร้างสัญญาณตัวอย่าง
N = 16  # จำนวนจุดข้อมูล
t = np.arange(N)
signal = np.sin(2 * np.pi * t / N) + 0.5 * np.sin(4 * np.pi * t / N)

# การคำนวณ DFT ด้วยนิยามตรงไปตรงมา
def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X

dft_result = dft(signal)

# การคำนวณ FFT ด้วย numpy
fft_result = np.fft.fft(signal)

# การแสดงผล
frequencies = np.arange(N)

plt.figure(figsize=(12, 6))

# สัญญาณต้นฉบับ
plt.subplot(1, 3, 1)
plt.stem(t, signal, basefmt=" ")
plt.title("Original Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")

# ผลลัพธ์ DFT
plt.subplot(1, 3, 2)
plt.stem(frequencies, np.abs(dft_result), basefmt=" ")
plt.title("DFT Result")
plt.xlabel("Frequency")
plt.ylabel("Magnitude")

# ผลลัพธ์ FFT
plt.subplot(1, 3, 3)
plt.stem(frequencies, np.abs(fft_result), basefmt=" ")
plt.title("FFT Result")
plt.xlabel("Frequency")
plt.ylabel("Magnitude")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# จำนวนจุดข้อมูล
N = 256
t = np.linspace(0, 1, N, endpoint=False)

# สร้างคลื่นแบบต่างๆ
sin_wave = np.sin(2 * np.pi * 5 * t)  # คลื่นไซน์ (Sine Wave)
square_wave = np.sign(np.sin(2 * np.pi * 5 * t))  # คลื่นสี่เหลี่ยม (Square Wave)
sawtooth_wave = 2 * (t * 5 - np.floor(t * 5 + 0.5))  # คลื่นฟันเลื่อย (Sawtooth Wave)
triangle_wave = 2 * np.abs(2 * (t * 5 - np.floor(t * 5 + 0.5))) - 1  # คลื่นสามเหลี่ยม (Triangle Wave)
random_noise = np.random.normal(0, 1, N)  # สัญญาณสุ่ม (Random Noise)

# ทำ FFT
fft_sin = np.fft.fft(sin_wave)
fft_square = np.fft.fft(square_wave)
fft_sawtooth = np.fft.fft(sawtooth_wave)
fft_triangle = np.fft.fft(triangle_wave)
fft_noise = np.fft.fft(random_noise)

# คำนวณความถี่
frequencies = np.fft.fftfreq(N, d=1/N)

# แสดงผลกราฟ
plt.figure(figsize=(12, 10))

# คลื่นไซน์
plt.subplot(5, 2, 1)
plt.plot(t, sin_wave)
plt.title("Sine Wave")
plt.grid(True)

plt.subplot(5, 2, 2)
plt.plot(frequencies[:N//2], np.abs(fft_sin[:N//2]))
plt.title("FFT of Sine Wave")
plt.grid(True)

# คลื่นสี่เหลี่ยม
plt.subplot(5, 2, 3)
plt.plot(t, square_wave)
plt.title("Square Wave")
plt.grid(True)

plt.subplot(5, 2, 4)
plt.stem(frequencies[:N//2], np.abs(fft_square[:N//2]), basefmt=" ")
plt.title("FFT of Square Wave")
plt.grid(True)

# คลื่นฟันเลื่อย
plt.subplot(5, 2, 5)
plt.plot(t, sawtooth_wave)
plt.title("Sawtooth Wave")
plt.grid(True)

plt.subplot(5, 2, 6)
plt.stem(frequencies[:N//2], np.abs(fft_sawtooth[:N//2]), basefmt=" ")
plt.title("FFT of Sawtooth Wave")
plt.grid(True)

# คลื่นสามเหลี่ยม
plt.subplot(5, 2, 7)
plt.plot(t, triangle_wave)
plt.title("Triangle Wave")
plt.grid(True)

plt.subplot(5, 2, 8)
plt.stem(frequencies[:N//2], np.abs(fft_triangle[:N//2]), basefmt=" ")
plt.title("FFT of Triangle Wave")
plt.grid(True)

# สัญญาณสุ่ม
plt.subplot(5, 2, 9)
plt.plot(t, random_noise)
plt.title("Random Noise")
plt.grid(True)

plt.subplot(5, 2, 10)
plt.stem(frequencies[:N//2], np.abs(fft_noise[:N//2]), basefmt=" ")
plt.title("FFT of Random Noise")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# กรองความถี่ที่เป็นบวก (positive frequencies)
positive_frequencies = frequencies[:N // 2]
positive_magnitudes = np.abs(fft_noise)[:N // 2]
print(positive_frequencies)
print(positive_magnitudes)
```

```python
# หาค่าความถี่ที่มี magnitude สูงสุด 2 อันดับแรก
top_two_indices = np.argsort(positive_magnitudes)[-2:]

# ความถี่ที่มี magnitude สูงสุด
top_frequencies = positive_frequencies[top_two_indices]
top_magnitudes = positive_magnitudes[top_two_indices]

print(top_two_indices)
print(top_frequencies)
print(top_magnitudes)
```

```python

# สร้างสัญญาณจากความถี่ที่มี magnitude สูงสุด
reconstructed_signal_1 = np.zeros(N)
reconstructed_signal_2 = np.zeros(N)

# ค้นหาดัชนีของความถี่ที่ต้องการใน FFT
freq_idx_1 = np.argmin(np.abs(positive_frequencies - top_frequencies[0]))
reconstructed_signal_1 += fft_noise[freq_idx_1].real * np.cos(2 * np.pi * positive_frequencies[freq_idx_1] * t)
reconstructed_signal_1 -= fft_noise[freq_idx_1].imag * np.sin(2 * np.pi * positive_frequencies[freq_idx_1] * t)

freq_idx_2 = np.argmin(np.abs(positive_frequencies - top_frequencies[1]))
reconstructed_signal_2 += fft_noise[freq_idx_2].real * np.cos(2 * np.pi * positive_frequencies[freq_idx_2] * t)
reconstructed_signal_2 -= fft_noise[freq_idx_2].imag * np.sin(2 * np.pi * positive_frequencies[freq_idx_2] * t)

# แสดงผลกราฟ
plt.figure(figsize=(12, 8))

# กราฟแสดงสัญญาณสุ่มต้นฉบับ
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 3)
plt.plot(t, random_noise)
plt.title("Original Random Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# กราฟแสดง magnitude ของ FFT เฉพาะด้านบวก
plt.subplot(3, 1, 1)
plt.plot(positive_frequencies, positive_magnitudes)
plt.title("Magnitude of FFT (Positive Frequencies)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# กราฟแสดงความถี่ที่มี magnitude สูงสุด 2 อันดับแรก
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 2)
plt.stem(top_frequencies, top_magnitudes, basefmt=" ")
plt.title("Top 2 Frequencies with Highest Magnitude (Positive Frequencies)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(t, reconstructed_signal_1, label=f"Reconstructed Signal at {top_frequencies[0]} Hz")
plt.title(f"Reconstructed Signals at {top_frequencies[0]} Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(t, reconstructed_signal_2, label=f"Reconstructed Signal at {top_frequencies[1]} Hz", linestyle='--')
plt.title(f"Reconstructed Signals at {top_frequencies[1]} Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
