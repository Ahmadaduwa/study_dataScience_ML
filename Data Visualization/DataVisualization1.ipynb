# Workshop 10: Graph with set details

```python
xs = [1,2,3,4,5,6]
ydata = [1,2.8,2.1,4,3,3.5]
ydata2 = [2.2,1.6,2.3,2,2.3,2.1]

plt.rcParams.update({"font.size":15, "font.family":"tahoma"})
plt.figure(figsize=(6,4)) #กว้าง สูง

plt.grid(linestyle="--", axis='y')
plt.bar(x, ydata, color='c', width=0.5, label='ในประเทศไทย')
plt.bar([x + 0.2 for x in xs],
        ydata2, width=0.5, color='orange', alpha=0.8, label='ส่งออก')

plt.plot(x, ydata, color='b', marker='s', markersize=8)
plt.plot(x, ydata2, color='r', marker='o', markersize=8)

plt.xlabel("เดือน")
plt.ylabel("ยอดขาย")
plt.title("รานงาน Sales Report")

for tx,ty in list(zip(x, ydata)):
    plt.annotate(ty,
                 (tx, ty),
                 textcoords='offset points',
                 xytext=(0,8),
                 ha='center',
                 fontsize=12)

plt.legend()
plt.ylim(0, 4.5)
plt.show()
```
